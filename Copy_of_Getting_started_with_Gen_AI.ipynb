```python
from langchain_core.prompts import ChatPromptTemplate

chat_template = ChatPromptTemplate.from_messages(
    [
        ("system", "You are a friendly assistant. Always respond with one clear, helpful one-line tip. Keep it encouraging and easy to follow."),
        ("human", "Share one simple tip about {topic} for beginners."),
    ]
)

messages = chat_template.format_messages(topic="staying hydrated")
messages
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite")
#from langchain_openai import ChatOpenAI
#llm = ChatOpenAI()
```

```python
ai_msg = llm.invoke(messages)
ai_msg
```

```python
from langchain_core.output_parsers import StrOutputParser
chain = chat_template | llm | StrOutputParser()
```

```python
chain.invoke({"topic": "sleeping better"})
```
